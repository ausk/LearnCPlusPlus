## Matplotlib refresher

```python
x = np.linspace(0, 2 * np.pi, 1000)
y0 = np.sin(x**3) / x
y1 = np.cos(x**2) * np.sin(x)

f, (ax0, ax1) = plt.subplots(1, 2, figsize=(10, 5))

ax0.plot(x, y0)
ax1.plot(x, y1);
```

```python
fig, ax = plt.subplots(figsize=(10, 5))

ax.plot(x, y0, linewidth=3, label=r'$\alpha$ signal')
ax.plot(x, y1, linewidth=3, label=r'$\beta$ signal')

ax.set_xlim(1, 6)         # or the global plt.xlim(1, 6)
ax.set_ylim(-1.2, 1.2)


plt.legend(fontsize=16, loc='lower center');
```

### Plotting regressions

```python
tips = sns.load_dataset("tips")
```

```python
sns.lmplot(x="total_bill", y="tip", data=tips);
```

```python
sns.lmplot(x="size", y="tip", data=tips);
```

Not so helpful with categorical variables!

```python
sns.lmplot(x="size", y="tip", data=tips, x_jitter=.05);
```

```python
sns.lmplot(x="size", y="tip", data=tips, x_estimator=np.mean);
```

```python
sns.lmplot("total_bill", "tip", tips, col="smoker");
```

### Plotting distributions

```python
x = np.random.normal(size=100)
sns.distplot(x);
```

```python
sns.distplot(x, kde=False, rug=True);
```

```python
from IPython.html.widgets import interact

@interact(bins=(5, 25, 5))
def show_distplot(bins):
    sns.distplot(x, bins=bins, kde=False, rug=True);
```

```python
sns.kdeplot(x)
sns.kdeplot(x, bw=.2, label="bw: 0.2", )
sns.kdeplot(x, bw=2, label="bw: 2")

plt.xlim(-4, 4)
plt.legend(fontsize=16);
```

```python
mean, cov = [0, 1], [(1, .5), (.5, 1)]
data = np.random.multivariate_normal(mean, cov, 200)
df = pd.DataFrame(data, columns=["x", "y"])
df.head()
```

```python
sns.jointplot(x="x", y="y", data=df);
```

```python
x, y = np.random.multivariate_normal(mean, cov, 1000).T
sns.jointplot(x=x, y=y, kind="hex");
```

```python
sns.jointplot(x="x", y="y", data=df, kind="kde");
```

## Facet Grids

``FacetGrid`` is used to draw plots with multiple
Axes where each Axes shows the same relationship conditioned on
different levels of some variable. It's possible to condition on up to
three variables by assigning variables to the rows and columns of the
grid and using different colors for the plot elements.

The general approach to plotting here is called "small multiples",
where the same kind of plot is repeated multiple times, and the
specific use of small multiples to display the same relationship
conditioned on one ore more other variables is often called a "trellis
plot".
